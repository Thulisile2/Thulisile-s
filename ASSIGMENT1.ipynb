```python
import numpy as np
import matplotlib.pyplot as plt


def Interest_monthtly(amount: float):  #Defining our first function
    return amount - (0.065 * amount)

def Method_To_Compute_Payments(amount: float):  #Defining our second function
    Interest = list([100_000])
    months = list([0])
    k = 0
    while True:
        balance = Interest[-1] - Interest_monthtly(amount)
        Interest.append(balance)
        k += 1
        months.append(k)
        if balance <= 0:
            break
       
    return months, Interest

x_1000, y_1000 = Method_To_Compute_Payments(1_000.00)
x_2000, y_2000 = Method_To_Compute_Payments(2_000.00)
x_3000, y_3000 = Method_To_Compute_Payments(3_000.00)
x_4000, y_4000 = Method_To_Compute_Payments(4_000.00)


#Plotting of a graph
plt.plot(x_1000, y_1000,label="1000")
plt.plot(x_2000, y_2000,label="2000")
plt.plot(x_3000,y_3000,label="3000")
plt.plot(x_4000, y_4000,label="4000")


#Labeling the axis and giving titles to our graph
plt.legend()
plt.ylim((0, 100_000.00))
plt.xlim((0, 60))
plt.title("The graph of balance over 5 years period")
plt.ylabel("debt")
plt.xlabel("No. of months")
axe=plt.gca()
plt.show()
```
